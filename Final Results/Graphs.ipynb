```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files
data1 = pd.read_csv("wc_without_prefault.csv")
data2 = pd.read_csv("wc_with_prefault.csv")
# data3 = pd.read_csv("C:\education\python virtual env 1\lkp-github\lkp-fault_around/random_access.csv")
for i in range(2, len(data1.columns)-1):
# Extract x values from the first column of data1
    x_values = data1.iloc[8:, 0]
    plt.figure(figsize=(10, 5))

    # Extract y values from the second column of both data1 and data2
    y1_values = data1.iloc[8:, i]
    y2_values = data2.iloc[8:, i]
    # y3_values = data3.iloc[:, i]
    # Plotting
    plt.plot(x_values, y1_values, label='No prefault')
    # plt.plot(x_values, y3_values, label='Data 2')
    plt.plot(x_values, y2_values, label='Prefault')
    plt.xlabel('Memory Size')
    plt.ylabel('Access Time (s)')
    plt.title('Order ' + str(i-1) )
    plt.yscale('log')
    plt.xticks(x_values, rotation=45)
    plt.legend()
    plt.grid(True)
    
plt.show()
```
